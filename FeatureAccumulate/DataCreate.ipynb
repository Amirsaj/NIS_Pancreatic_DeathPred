```python
def MLP_Res(X_train, X_test, y_train, y_true):
    MLPclf = MLPClassifier(random_state=42)
    MLPclf.fit(X_train, y_train)
    y_pred = MLPclf.predict(X_test)
    accuracy = accuracy_score(y_true, y_pred)
    f1metric =  f1_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred)
    recalll = recall_score(y_true, y_pred)
    return accuracy,f1metric,precision,recalll
```

```python
def RF_Res(X_train, X_test, y_train, y_true):
    from sklearn.ensemble import RandomForestClassifier

    RandomForestClassifier = RandomForestClassifier(max_depth= None, min_samples_leaf= 4, min_samples_split= 2, n_estimators= 200,random_state=42)
    RandomForestClassifier.fit(X_train, y_train)
    y_pred = RandomForestClassifier.predict(X_test)
    accuracy = accuracy_score(y_true, y_pred)
    f1metric =  f1_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred)
    recalll = recall_score(y_true, y_pred)
    return accuracy,f1metric,precision,recalll
```

```python
def SGD_Res(X_train, X_test, y_train, y_true):
    SGDclf = SGDClassifier(random_state=42)
    SGDclf.fit(X_train, y_train)
    y_pred = SGDclf.predict(X_test)
    accuracy = accuracy_score(y_true, y_pred)
    f1metric =  f1_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred)
    recalll = recall_score(y_true, y_pred)
    return accuracy,f1metric,precision,recalll
```

```python
def Ridge_Res(X_train, X_test, y_train, y_true):
    from sklearn.linear_model import RidgeClassifier

    ridgeclf = RidgeClassifier(random_state=42)
    ridgeclf.fit(X_train, y_train)
    y_pred = ridgeclf.predict(X_test)
    accuracy = accuracy_score(y_true, y_pred)
    f1metric =  f1_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred)
    recalll = recall_score(y_true, y_pred)
    return accuracy,f1metric,precision,recalll
```
